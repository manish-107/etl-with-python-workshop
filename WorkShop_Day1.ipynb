**Polymorphism**

````python
class Bird:
    def fly(self):
        print("Bird flies")

class Airplane:
    def fly(self):
        print("Airplane flies")

def make_it_fly(obj):
    obj.fly()

make_it_fly(Bird())
make_it_fly(Airplane())
````

**Composition (HAS-A Relationship)**

````python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()

    def drive(self):
        self.engine.start()
        print("Car moving")

c = Car()
c.drive()
````

**Custom Exceptions with Exception Object (as e)**

````python
class InsufficientBalanceError(Exception):
    pass

class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount > self.balance:
            raise InsufficientBalanceError("Not enough balance")
        self.balance -= amount


try:
    acc = BankAccount(1000)
    acc.withdraw(1500)
except InsufficientBalanceError as e:
    print(e)
````

````python
try:
    x = int(input("Enter number: "))
    y = int(input("Enter divisor: "))
    print(x / y)
except ValueError:
    print("ValueError: Invalid input")
except ZeroDivisionError:
    print("ZeroDivisionError: Cannot divide by zero")
````

**Create Student table**

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS student (
    student_id SERIAL PRIMARY KEY,
    roll_no VARCHAR(20) UNIQUE NOT NULL,
    name VARCHAR(100) NOT NULL
)
""")

conn.commit()
print("Student table created successfully")
cursor.execute("SELECT * FROM student")
print(cursor.fetchall())
````

**Create Subject table**

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS subject (
    subject_id SERIAL PRIMARY KEY,
    subject_code VARCHAR(20) UNIQUE NOT NULL,
    subject_name VARCHAR(100) NOT NULL
)
""")
conn.commit()
print("Subject table created successfully")
cursor.execute("SELECT * FROM subject")
print(cursor.fetchall())
````

**Marks Table with Foreign Keys**

````python
cursor.execute("""
CREATE TABLE IF NOT EXISTS marks (
    marks_id SERIAL PRIMARY KEY,
    student_id INT NOT NULL,
    subject_id INT NOT NULL,
    marks INT NOT NULL CHECK (marks >= 0 AND marks <= 100),

    CONSTRAINT fk_student
        FOREIGN KEY(student_id)
        REFERENCES student(student_id)
        ON DELETE CASCADE,

    CONSTRAINT fk_subject
        FOREIGN KEY(subject_id)
        REFERENCES subject(subject_id)
        ON DELETE CASCADE,

    CONSTRAINT unique_student_subject
        UNIQUE(student_id, subject_id)
)
""")
conn.commit()
print("Marks table created successfully")
cursor.execute("SELECT * FROM marks")
print(cursor.fetchall())
````

**Query All Student Marks**

````python
cursor.execute("""
SELECT s.name,
       sub.subject_name,
       m.marks
FROM marks m
JOIN student s ON m.student_id = s.student_id
JOIN subject sub ON m.subject_id = sub.subject_id
ORDER BY s.name
""")

rows = cursor.fetchall()
for row in rows:
    print(row)
````

**Query marks on one student**

````python
cursor.execute("""
SELECT sub.subject_name, m.marks
FROM marks m
JOIN subject sub ON m.subject_id = sub.subject_id
WHERE m.student_id = %s
""", (1,))

print(cursor.fetchall())
````
